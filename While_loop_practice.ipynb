```python
#11.Check whether a number is palindrome.
num = int(input('Enter  a number:'))
original = num
reverse = 0
while num > 0:
  digit = num % 10
  reverse = reverse * 10 + digit
  num = num // 10
if original == reverse:
  print('palindrome number')
else:
  print('not a palindrome number')
```

```python
#12. Print Fibonacci series using while.
n = int(input('Enter number:'))
a = 0
b = 1
i = 1
while i <= n:
  print(a)

  c = a + b
  a = b
  b = c

  i = i +1
```

```python
#14.Create a simple ATM menu using while.
balance = 1000
while True:
  print('\n----ATM MENU-------')
  print('1. check balance')
  print('2.deposit')
  print('3.whithdraw')
  print('4.Exit')
  choice = int(input('Enter your choice:'))
  if choice == 1:
    print('balance=',balance)

  elif choice == 2:
    amount =int(input('Enter deposit amount:'))
    balance = balance + amount
    print('Amount deposited successfully')

  elif choice == 3:
    amount=int(input('Enter withdrwal amount:'))
    if amount <= balance:
      balance = balance - amount
      print('please collect your cash')
    else:
      print('insufficient balance')


  elif choice == 4:
      print('Thank you for using ATM')
      break

  else:
      print('invalid choice')
```

```python
#17.Find greatest digit in a number.
num = int(input('Enter a number:'))
gretaest = 0
while num > 0:
  digit = num % 10
  if digit > gretaest:
    gretaest = digit
  num = num // 10
print('greatest digit = ',gretaest)
```

```python
#18.Find smallest digit in a number.
num = int(input('Enter a number:'))
smallest = 9
while num > 0:
  digit = num % 10
  if digit <  smallest:
    smallest = digit
  num = num // 10
print('smallest digit = ',smallest)
```
