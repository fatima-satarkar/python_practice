```python
#1.Print numbers from 1 to 20, but stop when the number becomes 10.
for i in range(1,21):
  print(i)
  if i == 10:
    break
```

```python
#2.Ask the user to enter numbers continuously and stop when they enter 0.
while True:
  num = int(input('Enter a number:'))
  if num == 0:
    break
```

```python
#3.Print numbers from 1 to 100 and stop when you find the first number divisible by 17.
for i in range(1,101):
  print(i)
  if i % 17 == 0:
    break
```

```python
#4.Search for a name in a list and stop searching once it is found.
names = ['apple','orange','grapes','strawberry']
for i in names:
  print(i)
  if i == 'orange':
    break
```

```python
#5.Print multiplication table of 5, but stop after 5 × 7.
for i in range(5,51,5):
  print(i)
  if i == 35:
    break
```

```python
#6.Keep asking for a password until the correct password is entered.
password = 'admin'
while password:
  password = input('Enter password:')
  if password == 'admin':
    break
```

```python
#7.Find the first even number in a list and stop the loop.
even = [2,5,6,7,8,9,0,13,4,5,6]
for i in even:
  print(i)
  if i %2==0:
    break
```

```python
#8.Print characters of a string and stop when a space is encountered.
a = 'python is a very beautiful language'
for i in a:
  print(i,end='')
  if i == ' ':
      break
```

```python
#9.Print numbers from 1 to 50 and stop when the running sum exceeds 100.
total = 0
for i in range(1,51):
  total = total + i
  print(i)

  if total > 100:
        break
```

```python
#10.Search for a vowel in a string and stop at the first vowel.
b = 'education'
for i in b:
  print(i)
  if i in 'aeiou':
    break
```
