```python
#6.Convert all names in a list to uppercase.
l = ['apple','orange','grapes','cherry']
upper_list = [i.upper()for i in l]
upper_list
```

```python
#6.Convert all names in a list to lowercase.
l = ['apple','orange','grapes','cherry']
lower_list = [i.lower()for i in l]
lower_list
```

```python
#9.Create a list of numbers divisible by 5 from 1 to 100.
[i for i in range(1,101) if i % 5 == 0]
```

```python
#10.Remove spaces from each word in a list.
l = ['apple  ','orange  ','grapes   ','cherry   ']
remove_space = [i.strip() for i in l]
remove_space
```

```python
#11.Create a list of vowels from a string.
s = 'education'
vowels = [i for i in s if i.lower() in 'aeiou']
vowels
```

```python
#18.Create a list of common elements from two lists.
ls1 = [1,2,3,4,5]
ls2 = [3,4,5,6,7]
common = [i for i in ls1 if i in ls2]
common
```

```python
#19.Flatten a nested list:
[[1,2],[3,4],[5,6]]
l = [[1,2],[3,4],[5,6]]
flatten = [j for i in l for j in i]
flatten
```

```python
#20.Replace negative numbers with 0 using list comprehension.
nums = [5,-2,8,-1,3,-7]
result = [i if i > 0 else 0 for i in nums]
result
```
