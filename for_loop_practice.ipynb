```python
#15.Check whether a number is palindrome.
num = input('Enter number:')
reverse = ''
for i in num:
  reverse = i +reverse
if num == reverse:
  print('Palindrome number')
else:
  print('not palindrome')
```

```python
#16.Check whether a number is Armstrong number.
num = input('Enter number:')
power = len(num)
total = 0
for i in num:
  total = total +int(i) ** power
if total == int(num):
  print('Armstrong number')
else:
  print('not armstrong number')
```

```python
#17.Print Fibonacci series up to n terms.
num = int(input('Enter numer of terms:'))
a = 0
b = 1
for i in range(num):
  print(a)

  c = a + b
  a = b
  b = c
```

```python
#18.Find greatest number in a list.
ls = [54,6,78,2,9,0]
greatest = ls[0]
for i in ls:
  if i > greatest:
    greatest = i
print(greatest)
```

```python

#19.Find smallest number in a list.
ls = [54,6,78,2,9,0]
smllatest = ls[0]
for i in ls:
  if i < smllatest:
    smllatest = i
print(smllatest)
```

```python
#20.Count vowels in a string.
text = 'education'
count = 0
for i in text:
  if i.lower() in 'aeiou':
    count = count + 1
print(count)
```

## **Pattern que**

```python
for i in range(1,6):
  for j in range(i):
    print('*',end='')
  print()
```

```python
for i in range(1,5):
  for j in range(1,i+1):
    print(j,end='')
  print()
```

```python
for i in range(5,0,-1):
  for j in range(i):
    print('*',end='')
  print()
```
