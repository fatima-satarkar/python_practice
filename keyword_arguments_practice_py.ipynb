```python
#1. Print Student Details
def student(name,age):
  print('name:',name)
  print('age:',age)
student(age=21,name='rahul')
```

```python
#2. Employee Details
def employee(name,salary,department):
  print('Name:',name)
  print('Salary:',salary)
  print('Department:',department)
employee(department='HR',name='anita',salary=50000)
```

```python
#3. Addition of Two Numbers
def add(a,b):
  print('sum=',a+b)
add(b= 20,a = 10)
```

```python
#4. Rectangle Area
def area(lenght,width):
  print('Area=',lenght + width)
area(width=5,lenght=10)
```

```python
#5. Mobile Information
def mobile(brand,model,price):
  print('Brand:',brand)
  print('Model:',model)
  print('Price:',price)
mobile(price=85000,brand='samsung',model='s25')
```

```python
#6. Book Details
def book(title,author):
  print('Title:',title)
  print('Author:',author)
book(author='john',title='python basics')
```

```python
#7. City and Country
def location(city,country):
  print(city,'is in',country)
location(country='India',city='mumbai')
```

```python
#8. Average Marks
def average(m1,m2,m3):
  print('Average=',(m1+m2+m3)/3)
average(m3=90,m1=80,m2=70)
```

```python
#9. Car Information
def car(brand,model,year):
  print('Brand:',brand)
  print('Model:',model)
  print('Year:',year)
car(year=2025,model='swift',brand='maruti')
```

```python
#10. Bank Account Details
def account(name,balance,acc_no):
  print('Name:',name)
  print('Balance:',balance)
  print('Account Number:',acc_no)
account(acc_no=12345,balance=50000,name='swapnil')
```

```python
#13. Product Details
def product(name,price,quantity):
  print('Name:',name)
  print('price:',price)
  print('Quantity:',quantity)
product(price=45,name='alo gel',quantity=2)
```

```python
#14. Calculate Simple Interest
def interest(p,r,t):
  print('Simple Interest=',(p*r*t)/100)
interest(t=9,r=4,p=2)
```
